```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
tr = csv.reader(open('tr.csv', 'r'), delimiter=',')
val = csv.reader(open('val.csv', 'r'), delimiter=',')
```

```python
cnt = 0
tr_dice = []
for row in tr:
    if cnt == 0:
        cnt+=1
        continue
    dice = float(row[2])
    tr_dice.append(dice)
```

```python
cnt = 0
val_dice = []
for row in val:
    if cnt == 0:
        cnt+=1
        continue
    dice = float(row[2])
    val_dice.append(dice)
```

```python
tr_dice = np.array(tr_dice)
```

```python
val_dice = np.array(val_dice)
```

```python
epochs = np.arange(0, len(val_dice), 1)
```

```python
plt.rcParams.update({'font.size': 18, 'legend.loc':'center right'})
plt.figure(figsize=(10,7))
plt.plot(epochs, tr_dice)
plt.plot(epochs, val_dice, 'r')
plt.title('Dice Coefficient vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Dice Coefficient')
plt.legend(['Trianing', 'Validation'])
plt.ylim([0.5, 0.75])
plt.xlim([0, 300])
plt.grid()
plt.savefig('seg_tr_curv.png')
plt.show()
```
